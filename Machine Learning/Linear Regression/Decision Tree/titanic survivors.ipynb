<h1 align='center'>Decision Tree Classification</h1>

<center>We have a data for Titanic where we have to predict that a weather a person would survive or not

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

### We'll drop all the column that are not useful inorder to predict the survivor

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns', inplace=True)
df.head()
```

### Now survived is the result so we have set it to target and the remainings are set to independent value called inputs

```python
inputs = df.drop(['Survived'], axis='columns')
target = df.Survived
```

### We'll transform the labels in number as Machine Learning Model can only work with numbers and not with labels

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs
```

### Now to fill the NaN in Age column. We will take a mean of the age column and would replace that with all the NaN values in Age Column

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

### As we have gendar in numbers so we can drop gendar column which is in text

```python
inputs.drop(['Sex'], axis='columns', inplace=True)
```

```python
inputs.head()
```

### Now we'll train our model with 80% training data and 20% testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
